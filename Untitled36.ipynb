```python
import cv2
import numpy as np
import time
from google.colab.patches import cv2_imshow
```

```python

# Load YOLO
net = cv2.dnn.readNet("/content/yolov3-tiny.weights", "/content/raw.githubusercontent.com_pjreddie_darknet_master_cfg_yolov3-tiny.cfg.txt")
```

```python

# Enable GPU acceleration (if available)
net.setPreferableBackend(cv2.dnn.DNN_BACKEND_CUDA)
net.setPreferableTarget(cv2.dnn.DNN_TARGET_CUDA)
```

```python
# Define the classes
classes = []
with open("/content/raw.githubusercontent.com_pjreddie_darknet_master_data_coco.names.txt", "r") as f:
    classes = [line.strip() for line in f.readlines()]
```

```python
# Function to perform object detection
def detect_objects(image):
    # Create a blob from the image
    blob = cv2.dnn.blobFromImage(image, 1/255.0, (416, 416), swapRB=True, crop=False)

    # Set the input to the network
    net.setInput(blob)

    # Run forward pass and get the output
    start_time = time.time()
    outputs = net.forward(net.getUnconnectedOutLayersNames())
    end_time = time.time()

    # Get the dimensions of the image
    height, width, _ = image.shape

    # Process the output detections
    for output in outputs:
        for detection in output:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]

            if confidence > 0.3:
                # Calculate the bounding box coordinates
                box = detection[0:4] * np.array([width, height, width, height])
                (center_x, center_y, box_width, box_height) = box.astype("int")

                # Calculate the top-left corner coordinates
                x = int(center_x - (box_width / 2))
                y = int(center_y - (box_height / 2))

                # Draw the bounding box and label on the image
                cv2.rectangle(image, (x, y), (x + box_width, y + box_height), (0, 255, 0), 2)
                cv2.putText(image, str(class_id), (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

                # Print the bounding box coordinates
                print("Bounding Box:", x, y, box_width, box_height)

    # Print the time taken for object detection
    print("Object detection time: {:.2f} seconds".format(end_time - start_time))

    # Display the result
    cv2_imshow(image)

# Load the image
image_path = "/content/009960.jpg"
image = cv2.imread(image_path)

# Perform object detection
detect_objects(image)
```
